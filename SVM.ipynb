### Import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### Import dataset

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df.info()
```

```python
df.Gender.value_counts()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(df.Gender)
plt.show()
```

```python
# Dependent Fearures
X = df[["EstimatedSalary","Age"]]
# Independent Features
Y = df["Purchased"]
```

```python
X.head()
```

```python
Y.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size=0.25,random_state=20)
```

```python
print("Training Data = ",X_Train.shape)
print("Testing Data = ",X_Test.shape)
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Train = scaler.fit_transform(X_Train)
X_Test = scaler.transform(X_Test)
```

### Model Building

#### 1. Using Linear Kernel

```python
from sklearn.svm import SVC
model = SVC(kernel="linear", random_state=0)
model.fit(X_Train,Y_Train)

## Prediction
Y_Pred = model.predict(X_Test)
Y_Pred 
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_Test,Y_Pred)
accuracy
```

#### 2.Using Polynomial Kernel

```python
from sklearn.svm import SVC
model= SVC(kernel="poly",degree =3, random_state=0)
model.fit(X_Train,Y_Train)

## Prediction
Y_Pred = model.predict(X_Test)
Y_Pred 
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_Test,Y_Pred)
accuracy
```

#### 3. Using RBF Kernel

```python
from sklearn.svm import SVC
model= SVC(kernel="rbf",random_state=0)
model.fit(X_Train,Y_Train)

## Prediction
Y_Pred = model.predict(X_Test)
Y_Pred 
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_Test,Y_Pred)
accuracy
```

```python
## With Regularization=C and Gamma values
from sklearn.svm import SVC
# model = SVC(kernel="rbf",gamma=15,C=7, random_state=0)
model = SVC(kernel="rbf",gamma=1,C=50,random_state=0)
model.fit(X_Train,Y_Train)

## Prediction
Y_Pred = model.predict(X_Test)
Y_Pred 
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_Test,Y_Pred)
accuracy
```

### Why RBF Kernel Giving Good Accuracy?
- Becuase, linear kernel does good job when data is linearaly seperable
- Polynomial kernel does good job when data is seperable in N-dimensional space
- But RBF kernel works well when data is linearly not seperable, and since our data is linear non seperable hence RBF giving us good accuracy ober Polynomial and linear kernel
